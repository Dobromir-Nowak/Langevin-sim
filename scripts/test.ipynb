```python
from langevin_sim.utils import plot_trajectories, load_config
from langevin_sim.geometry import Cylinder3D
import numpy as np
```

```python
file_name = "config_Cylinder3D"
config = load_config(file_name=file_name)

geometry = Cylinder3D(config=config)
```

```python
r_old = np.array([3,4,0])[:, None]
r_new = np.array([15, 20, 0])[:, None]
r_reflection = np.copy(r_old)
geometry.apply(r_reflection,r_new)

r_old, r_reflection, r_new = r_old[None,:], r_reflection[None,:], r_new[None,:]
r_history = np.concatenate([r_old,r_reflection,r_new], axis=0)
print(r_history)
# r_history.shape, r_reflection.shape, r_new.shape, r_old.shape
plot_trajectories(r_history= r_history, n_show=r_history.shape[2])
```
